```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nwrepo = pd.read_csv('results/runtime-networkrepository.csv')
nwrepo['source'] = 'networkrepository'
konect = pd.read_csv('results/runtime-konect.csv')
konect['source'] = 'konect'

data = pd.concat([nwrepo, konect])
data.groupby('source').file.nunique()
```

```python
colors = {'pq': 'red', 'set': 'blue', 'treap': 'green'}

for source, source_view in data.groupby('source'):
    print(source)
    ax = plt.gca()
    plt.yscale('symlog')
    for algo, view in source_view.groupby('algo'):
        view.plot.scatter('n', 'time', ax=ax, label=algo, logx=True, c=colors[algo])
    plt.show()
        
    source_view.groupby('algo').time.sum().plot.bar()
    plt.show()
```
